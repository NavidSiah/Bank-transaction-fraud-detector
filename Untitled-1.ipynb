```python
import pandas as pd

dataset_path = 'creditcard.csv'
data = pd.read_csv(dataset_path)
```

```python
# Displaying the data structure (number of rows and columns)
print("number of rows and columns:")
print(data.shape)
print("\n\n")

# Display column names
print("Dicolumn names:")
print(data.columns)
print("\n\n")


# Display the type of data in each column
print("type of data in each column:")
print(data.dtypes)
print("\n\n")


# Display various statistics(number, average, standard deviation, etc.)
print("various statistics:")
print(data.describe())
print("\n\n")


# Display the number of unique values ​​in each column
print("number of unique values ​​in each column")
print(data.nunique())
print("\n\n")
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt

normalized_data = data.copy()

scaler = MinMaxScaler()
coloumns_to_normalize = data.columns
normalized_data[coloumns_to_normalize] = scaler.fit_transform(data[coloumns_to_normalize])

class_data = normalized_data['Class']
normalized_data = normalized_data.drop(['Class'], axis=1)

number_of_clusters = 2

kmeans = KMeans(n_clusters=number_of_clusters, random_state=0)
kmeans.fit(normalized_data)

labels = kmeans.labels_

cluster_samples0 = normalized_data[labels == 0]
cluster_samples1 = normalized_data[labels == 1]


frauds0 = 0
notFrauds0 = 0
frauds1 = 0
notFrauds1 = 0
for i in range(len(labels)):
    if labels[i] == 0:
        if class_data[i] == 0:
            notFrauds0 += 1
        else:
            frauds0 += 1
    else:
        if class_data[i] == 0:
            notFrauds1 += 1
        else:
            frauds1 += 1

class_counts = data['Class'].value_counts()
print(class_counts)

print("number of samples in cluster 0 : ", len(cluster_samples0))
print("number of frauds in cluster 0 : ", frauds0)

print("number of samples in cluster 1 : ", len(cluster_samples1))
print("number of frauds in cluster 1 : ", frauds1)

accuracy = accuracy_score(labels, data['Class'])
print("accuracy : ", accuracy)


plt.pie([frauds0, notFrauds0], labels=['fraud', 'notFraud'], autopct='%1.1f%%')
plt.title('cluster 0')
plt.show()


plt.pie([frauds1, notFrauds1], labels=['fraud', 'notFraud'], autopct='%1.1f%%')
plt.title('cluster 1')
plt.show()
```

```python
number_of_clusters = 143

kmeans = KMeans(n_clusters=number_of_clusters, random_state=0)
kmeans.fit(normalized_data)

labels = kmeans.labels_

frauds = [0 for i in range(number_of_clusters)]
for i in range(len(labels)):
    if class_data[i] == 1:
        frauds[labels[i]] += 1
        
for i in range(number_of_clusters):
    print("number of samples in cluster ", i, ": ", len(normalized_data[labels == i]))

for i in range(number_of_clusters):
    print(i , (frauds[i] / len(normalized_data[labels == i])) * 100)

for i in range(len(frauds)):
    print(i, frauds[i])
```

```python
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE

scaler1 = MinMaxScaler()

Y = data['Class']
X = data.drop(['Class'], axis=1)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

smote = SMOTE(sampling_strategy=0.05)
X_train, Y_train = smote.fit_resample(X_train, Y_train)

X_train = scaler1.fit_transform(X_train)
X_test = scaler1.transform(X_test)

class_counts = data['Class'].value_counts()
labels = ['Class 0', 'Class 1']
plt.pie(class_counts, labels=labels, autopct='%1.1f%%')
plt.title(f'Number of Classes\n0: {class_counts[0]} \n1: {class_counts[1]}')
plt.show()

class_counts = Y_train.value_counts()
labels = ['Class 0', 'Class 1']
plt.pie(class_counts, labels=labels, autopct='%1.1f%%')
plt.title(f'Number of Classes\n0: {class_counts[0]} \n1: {class_counts[1]}')
plt.show()

class_counts = Y_test.value_counts()
labels = ['Class 0', 'Class 1']
plt.pie(class_counts, labels=labels, autopct='%1.1f%%')
plt.title(f'Number of Classes\n0: {class_counts[0]} \n1: {class_counts[1]}')
plt.show()
```

```python
from sklearn import svm
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
import seaborn as sns


model = svm.SVC()

model.fit(X_train, Y_train)

Y_pred = model.predict(X_test)

accuracy = accuracy_score(Y_test, Y_pred)

print("Accuracy:", accuracy)

cm = confusion_matrix(Y_test, Y_pred)

sns.heatmap(cm, annot=True, fmt='d', cmap='YlGn')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
report = classification_report(Y_test, Y_pred)

print(report)
```

```python
from sklearn.linear_model import LogisticRegression


model = LogisticRegression()

model.fit(X_train, Y_train)

y_pred = model.predict(X_test)

accuracy = accuracy_score(Y_test, y_pred)

print("Accuracy:", accuracy)

print(classification_report(Y_test, y_pred))

cm = confusion_matrix(Y_test, y_pred)

sns.heatmap(cm, annot=True, fmt='d', cmap='YlGn')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```
